```python
# Create recommended features dataset from cleaned_UA.csv
import pandas as pd

in_path = "../data/processed/cleaned_UA.csv"
out_path = "../data/processed/reccomended_features_dataset.csv"

# Select the desired columns — note 'Smoking' and 'Drinking' are separate, and target is 'Fracture'
selected = [
    "Gender",
    "Age",
    "Height",
    "Weight",
    "BMI",
    "Smoking",
    "Drinking",
    "Fracture",
]

# Load cleaned data
df_clean = pd.read_csv(in_path)

# Check for missing columns and report
missing = [c for c in selected if c not in df_clean.columns]
if missing:
    print(f"Warning: the following requested columns are missing in {in_path}: {missing}")

rec_df = df_clean[[c for c in selected if c in df_clean.columns]].copy()

# Save recommended dataset
rec_df.to_csv(out_path, index=False)
print(f"Saved {rec_df.shape[0]} rows x {rec_df.shape[1]} cols to {out_path}")

# Display a quick preview
rec_df.head()
```

```python
# Split `reccomended_features_dataset.csv` into 90/10 train/test and save (robust target detection)
import pandas as pd

# Ensure scikit-learn is available in the notebook kernel; install if missing
try:
    from sklearn.model_selection import train_test_split
except ModuleNotFoundError:
    import sys, subprocess
    print("scikit-learn not found in kernel; installing into kernel environment...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "scikit-learn"]) 
    from sklearn.model_selection import train_test_split

in_path = "../data/processed/reccomended_features_dataset.csv"
train_path = "../data/processed/train_reccomended_features.csv"
test_path = "../data/processed/test_reccomended_features.csv"

# Load the recommended dataset
df_rec = pd.read_csv(in_path)
print("Loaded rec dataset:", df_rec.shape)
print("Columns:", list(df_rec.columns))

# Robust target finder: match column names containing 'fracture' (ignore case/spaces/underscores)
def find_target(cols, name="fracture"):
    name_clean = name.lower().replace(" ", "").replace("_", "")
    for c in cols:
        c_clean = c.lower().replace(" ", "").replace("_", "")
        if name_clean in c_clean:
            return c
    return None

target_col = find_target(df_rec.columns, "fracture")
if target_col is None:
    raise ValueError(f"Target column containing 'fracture' not found. Available columns: {list(df_rec.columns)}")

print(f"Using target column: {target_col}")

# Drop rows with missing target
missing_target = df_rec[target_col].isnull().sum()
if missing_target:
    print(f"Dropping {missing_target} rows with missing target '{target_col}'")
    df_rec = df_rec[df_rec[target_col].notnull()].copy()

X = df_rec.drop(columns=[target_col])
y = df_rec[target_col]

# Determine stratify param only when there are >=2 classes
stratify_param = y if y.nunique() > 1 else None

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, stratify=stratify_param, random_state=42
)

train_df = pd.concat([X_train, y_train], axis=1)
test_df = pd.concat([X_test, y_test], axis=1)

train_df.to_csv(train_path, index=False)
test_df.to_csv(test_path, index=False)

print(f"Saved train: {train_df.shape} -> {train_path}")
print(f"Saved test:  {test_df.shape} -> {test_path}")

train_df.head()
```

```python
# EDA plots for the recommended features dataset
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_path = "../data/processed/train_reccomended_features.csv"
plots_dir = "../outputs/plots"
os.makedirs(plots_dir, exist_ok=True)

# Load training data only for EDA to avoid test leakage
eda_df = pd.read_csv(train_path)

# Ensure the expected columns exist
expected_cols = ["Gender", "Age", "Height", "Weight", "BMI", "Smoking", "Drinking", "Fracture"]
missing_cols = [col for col in expected_cols if col not in eda_df.columns]
if missing_cols:
    raise ValueError(f"Missing columns in {train_path}: {missing_cols}")

# Force numeric values for plotting
for col in ["Age", "Height", "Weight", "BMI", "Smoking", "Drinking", "Fracture"]:
    eda_df[col] = pd.to_numeric(eda_df[col], errors="coerce")

sns.set_theme(style="whitegrid", context="talk")

saved_plots = []

def save_current_figure(filename):
    full_path = os.path.join(plots_dir, filename)
    plt.savefig(full_path, dpi=200, bbox_inches="tight")
    saved_plots.append(full_path)
    plt.close()

# 1) Target class balance
fig, ax = plt.subplots(figsize=(7, 5))
fracture_order = sorted(eda_df["Fracture"].dropna().unique().tolist())
sns.countplot(data=eda_df, x="Fracture", order=fracture_order, color="#4c72b0", ax=ax)
ax.set_title("Fracture Class Balance")
ax.set_xlabel("Fracture")
ax.set_ylabel("Count")
for container in ax.containers:
    ax.bar_label(container)
save_current_figure("01_fracture_class_balance.png")

# 2) Age distribution
fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(eda_df["Age"].dropna(), bins=30, kde=True, color="#2a9d8f", ax=ax)
ax.set_title("Age Distribution")
ax.set_xlabel("Age")
ax.set_ylabel("Count")
save_current_figure("02_age_distribution.png")

# 3) BMI distribution
fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(eda_df["BMI"].dropna(), bins=30, kde=True, color="#e76f51", ax=ax)
ax.set_title("BMI Distribution")
ax.set_xlabel("BMI")
ax.set_ylabel("Count")
save_current_figure("03_bmi_distribution.png")

# 4) Height distribution
fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(eda_df["Height"].dropna(), bins=30, kde=True, color="#457b9d", ax=ax)
ax.set_title("Height Distribution")
ax.set_xlabel("Height")
ax.set_ylabel("Count")
save_current_figure("04_height_distribution.png")

# 5) Weight distribution
fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(eda_df["Weight"].dropna(), bins=30, kde=True, color="#8d99ae", ax=ax)
ax.set_title("Weight Distribution")
ax.set_xlabel("Weight")
ax.set_ylabel("Count")
save_current_figure("05_weight_distribution.png")

# 6) Correlation heatmap
fig, ax = plt.subplots(figsize=(10, 8))
correlation_cols = ["Age", "Height", "Weight", "BMI", "Smoking", "Drinking", "Fracture"]
corr = eda_df[correlation_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0, square=True, linewidths=0.5, ax=ax)
ax.set_title("Correlation Heatmap")
save_current_figure("06_correlation_heatmap.png")

# 7) Fracture vs Smoking/Drinking grouped bar chart
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)
feature_hues = [("Smoking", "Fracture vs Smoking"), ("Drinking", "Fracture vs Drinking")]
for axis, (hue_col, title) in zip(axes, feature_hues):
    hue_order = sorted(eda_df[hue_col].dropna().unique().tolist())
    sns.countplot(
        data=eda_df,
        x="Fracture",
        hue=hue_col,
        order=fracture_order,
        hue_order=hue_order,
        palette="Set2",
        ax=axis,
    )
    axis.set_title(title)
    axis.set_xlabel("Fracture")
    axis.set_ylabel("Count")
    axis.legend(title=hue_col, loc="upper right")
fig.suptitle("Fracture vs Smoking/Drinking Grouped Bar Charts", y=1.02)
fig.tight_layout()
plt.savefig(os.path.join(plots_dir, "07_fracture_vs_smoking_drinking_grouped_bars.png"), dpi=200, bbox_inches="tight")
saved_plots.append(os.path.join(plots_dir, "07_fracture_vs_smoking_drinking_grouped_bars.png"))
plt.close(fig)

print("Saved plots:")
for plot_path in saved_plots:
    print(plot_path)
```
